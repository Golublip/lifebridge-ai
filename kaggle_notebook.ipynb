# LifeBridge AI - Clinical Symptom Classification & Patient Adherence Modeling

Welcome to the **LifeBridge AI** Kaggle notebook! This notebook walks through data pre-processing, clinical symptom intent classification, and medication adherence trend analysis. This models the core machine learning underpinnings of the LifeBridge AI healthcare navigator.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

print("Libraries loaded successfully.")
```

## 1. Symptom Intent Classification Model

We construct a dataset of patient messages mapped to core intents (`emergency`, `medication`, `symptom`, `hospital`, `nutrition`, `mental`).

```python
# Create sample data representing patient inputs
data = [
    ("I have severe chest pressure and left arm pain", "emergency"),
    ("My face is drooping and my speech is slurred", "emergency"),
    ("How should I schedule my Metformin tablets?", "medication"),
    ("Can I take Lisinopril with Potassium supplements?", "medication"),
    ("I have a mild cough, runny nose, and light fever", "symptom"),
    ("Where can I find a low-cost diabetes clinic?", "hospital"),
    ("Are there government sliding scale clinics near 43102?", "hospital"),
    ("What foods are best for managing type 2 diabetes?", "nutrition"),
    ("I feel extremely anxious and stressed about my health", "mental"),
    ("Can you give me a mindfulness exercise to calm down?", "mental")
] * 10  # Duplicate to simulate larger training set

df = pd.DataFrame(data, columns=["text", "intent"])

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(df["text"], df["intent"], test_size=0.2, random_state=42)

# Train simple model
model = Pipeline([
    ('tfidf', TfidfVectorizer(stop_words='english')),
    ('clf', LogisticRegression(C=10.0))
])

model.fit(X_train, y_train)
preds = model.predict(X_test)

print(classification_report(y_test, preds))
```

## 2. Patient Vitals Trend Visualization

We model fasting blood sugar and systolic blood pressure logs for diabetic and hypertensive patients over time.

```python
# Generate mock patient data
np.random.seed(42)
days = np.arange(1, 31)
glucose_values = 135 + np.random.normal(0, 15, size=30)  # Elevated levels
bp_systolic = 130 + np.random.normal(0, 5, size=30)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(days, glucose_values, color='tab:blue', marker='o', label='Fasting Glucose')
plt.axhline(100, color='tab:green', linestyle='--', label='Normal limit')
plt.title('Fasting Glucose Trend (30 Days)')
plt.xlabel('Day')
plt.ylabel('Glucose (mg/dL)')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(days, bp_systolic, color='tab:red', marker='s', label='Systolic BP')
plt.axhline(120, color='tab:green', linestyle='--', label='Prehypertensive limit')
plt.title('Systolic BP Trend (30 Days)')
plt.xlabel('Day')
plt.ylabel('BP (mmHg)')
plt.legend()

plt.tight_layout()
plt.show()
```
